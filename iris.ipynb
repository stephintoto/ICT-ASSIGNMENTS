```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')

print(iris.head())
```

violin plot:

The violin plot shows that setosa has a narrower range of sepal lengths compared to versicolor and virginica. This indicates less variability in setosa.

```python

plt.figure(figsize=(5, 5))
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.title('Violin Plot of Sepal Length by Species')
plt.show()
```

swarm plot :shows the distribution of petal widths, showing that setosa has significantly smaller petal widths than the other species, while versicolor and virginica overlap somewhat but still show distinguishable patterns.

```python

plt.figure(figsize=(5, 5))
sns.swarmplot(x='species', y='sepal_length', data=iris)
plt.title('Swarm Plot of Sepal Length by Species')
plt.show()

```

The strip plot with jitter adds clarity to the distribution of sepal lengths within each species, particularly showing how setosa is distinctly separated from the others.

```python

plt.figure(figsize=(5, 5))
sns.stripplot(x='species', y='sepal_length', data=iris, jitter=True)
plt.title('Strip Plot of Sepal Length by Species')
plt.show()
```

The box plot highlights that setosa generally has higher sepal widths compared to versicolor and virginica. Outliers in versicolor and virginica indicate variability in sepal width.

```python

plt.figure(figsize=(5, 5))
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.title('Box Plot of Sepal Length by Species')
plt.show()
```

The bar plot shows that virginica has the longest average petal length, followed by versicolor and then setosa, which has the shortest petals.

```python

plt.figure(figsize=(5, 5))
sns.barplot(x='species', y='sepal_length', data=iris)
plt.title('Bar Plot of Sepal Length by Species')
plt.show()
```

 The count plot confirms that the Iris dataset is balanced, with 50 samples for each species.

```python

plt.figure(figsize=(8, 6))
sns.countplot(x='species', data=iris)
plt.title('Count Plot of Species')
plt.show()
```

The scatterplot shows that setosa has a unique cluster of points with shorter sepal lengths and wider sepal widths, while versicolor and virginica overlap more but still show distinct patterns.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris, palette='deep')
plt.title('Scatterplot of Sepal Length vs. Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

The line chart shows the variation in sepal length across the dataset. You can observe the clear differences in sepal length between species, with setosa consistently having the shortest sepals and virginica the longest.

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=iris.index, y='sepal_length', hue='species', data=iris, marker='o')
plt.title('Line Chart of Sepal Length across Observations')
plt.xlabel('Observation Index')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

The heatmap reveals strong correlations between petal_length and petal_width, suggesting that these features could be redundant in some analyses.

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
corr_matrix = iris.drop(columns=['species']).corr()
print(corr_matrix)

plt.figure(figsize=(5, 5))

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

plt.title('Correlation Matrix of Iris Dataset')

plt.show()

```
