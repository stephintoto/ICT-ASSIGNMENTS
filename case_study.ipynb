```python
uniq_workclass= df['workclass'].unique()
uniq_workclass
```

```python
df['income_binary'] = df['income'].apply(lambda x: 1 if x == '>50K' else 0)
print(df['income_binary'].head())
```

```python
df['age_group'] = df['age'].apply(lambda x: 'Youth' if x < 30 else ('Senior' if x >= 60 else 'Adult'))
print(df['age_group'].head())
```

```python
df.rename(columns={'educational-num': 'education_num'}, inplace=True)
print(df.columns)
```

```python
df['workclass'].replace('?', 'Unknown', inplace=True)
df['occupation'].replace('?', 'Unknown', inplace=True)
print(df[['workclass', 'occupation']].head())
```

```python
avgage_workclass = df.groupby('workclass')['age'].mean().reset_index()
print(avgage_workclass)
```

```python
df['capital-gain'].replace('?', 0, inplace=True)
df['capital-gain'] = pd.to_numeric(df['capital-gain'], errors='coerce')
gendergain= df.groupby('gender')['capital-gain'].sum().reset_index()
print(gendergain)
```

```python
marital_status_counts = df['marital-status'].value_counts().reset_index()
marital_status_counts.columns = ['marital_status', 'count']
print(marital_status_counts)
```

```python
edu_mean = df.groupby('education')['hours-per-week'].mean().reset_index()
print(edu_mean)
```

```python
fnlwgt_stat= df.groupby('race')['fnlwgt'].agg(['mean', 'median', 'std']).reset_index()
print(fnlwgt_stat)
```

```python
country_counts = df['native-country'].value_counts().reset_index()
country_counts.columns = ['native_country', 'count']
print(country_counts)
```

```python
filtered_dataset = df[df['income'] == '>50K']
filtered_dataset
```

```python
sorted_dataset = df.sort_values(by='age', ascending=False)
print(sorted_dataset)
```

```python
hr_week = df[df['hours-per-week'] > 40]
print(hr_week)
```

```python
missing_percentage = (df.isnull().sum() / len(df)) * 100
miss_percen_df = missing_percentage.reset_index()
miss_percen_df.columns = ['column', 'missing_percentage']
miss_percen_df
```

```python
income_mapping = {
    '>50K': 1,
    '<=50K': 0
}
df['income_numeric'] = df['income'].map(income_mapping)
median_income= df.groupby(['education', 'gender'])['income_numeric'].median().reset_index()
median_income.columns = ['education', 'gender', 'median_income']
print(median_income)
```

```python
top3_occupations = df.groupby('occupation')['capital-gain'].mean().nlargest(3).reset_index()
print(top3_occupations)
```
