````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

diabetes_df = pd.read_csv('/content/diabetes_model.csv')

fig, axes = plt.subplots(9, 2, figsize=(15, 30))

for i, column in enumerate(diabetes_df.columns):
    sns.histplot(diabetes_df[column], kde=True, ax=axes[i, 0])
    axes[i, 0].set_title(f'Histogram of {column}')

    sns.boxplot(x=diabetes_df[column], ax=axes[i, 1])
    axes[i, 1].set_title(f'Boxplot of {column}')

plt.tight_layout()
plt.show()
````

````python

corr_matrix = diabetes_df.corr()

sns.pairplot(diabetes_df, hue='Outcome', diag_kind='kde', markers=['o', 's'])

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
````
